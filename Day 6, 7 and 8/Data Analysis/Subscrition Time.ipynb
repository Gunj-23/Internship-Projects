```python
import pandas as pd
import matplotlib.pyplot as plt

# Convert to datetime if not already
df['Subscription Date'] = pd.to_datetime(df['Subscription Date'])

# Group by month (or day, or year) and count subscriptions
subscriptions_per_month = df.groupby(df['Subscription Date'].dt.to_period('M')).size()

# Convert PeriodIndex to datetime for plotting
subscriptions_per_month.index = subscriptions_per_month.index.to_timestamp()

# Plot
plt.figure(figsize=(12,6))
subscriptions_per_month.plot(kind='line', marker='o')
plt.title('Number of Subscriptions per Month')
plt.xlabel('Month')
plt.ylabel('Number of Subscriptions')
plt.grid(True)
plt.show()
```
